```python
import pandas as pd
import matplotlib.pyplot as plt
#import seaborn as sns
import numpy as np 
import csv
import json
import pprint as pprint
```

```python
temp_data = pd.read_csv("../bee_data/city_temperature.csv", low_memory =False)
temp_data
```

```python
temp_data.drop(temp_data.loc[temp_data['Country'] != 'US'].index, inplace=True)
temp_data
```

```python
temp_data.dtypes
```

```python
## Checking for outliers in the temperature data set
temp_check = temp_data.plot.scatter(x='Year',
                      y='AvgTemperature',
                      c='DarkBlue')
```

```python
temp_data.drop(temp_data.loc[temp_data['AvgTemperature'] == -99.00].index, inplace=True)
temp_data
```

```python
state_avg_temp = temp_data.groupby(['State','Month','Day','Year'], as_index = False).mean(['AvgTemperature'])
state_avg_temp
```

```python
state_avg_temp.dtypes
```

```python
state_avg_temp_filtered = state_avg_temp[(state_avg_temp['AvgTemperature']<= 41) | (state_avg_temp['AvgTemperature']>= 95)]
```

```python
state_avg_temp_filtered
```

```python
state_avg_temp_filtered.to_csv('Test.csv')
```

```python
state_count = state_avg_temp_filtered.groupby(['State','Year'], as_index = False)['AvgTemperature'].count()
state_count
```

```python
cleaned_state_count = state_count.rename(columns={'Year':'year','State':'state','AvgTemperature':'extreme_temp_days'})
cleaned_state_count
```

```python
cleaned_state_count_subset =  cleaned_state_count[(cleaned_state_count['year']>= 2015)]
cleaned_state_count_subset
```

```python
us_total_temp = cleaned_state_count_subset.groupby(['year'], as_index = False).median(['extreme_temp_days'])
us_total_temp
```

```python
us_total_temp['state']="US TOTAL"
us_total_temp
```

```python
state_2018_ext_temp =  cleaned_state_count_subset[(cleaned_state_count_subset['year']>= 2018)]
```

```python
extreme_temp_data = pd.concat([us_total_temp, state_2018_ext_temp], join="inner")
extreme_temp_data
```
